```python
from openai import OpenAI

client = OpenAI(
    api_key="",
    # With the standard OpenAI client, the /openai/v1 suffix is REQUIRED
    base_url="https://smartde.openai.azure.com/openai/v1"
)
```

```python
research_paper_text = """
The Smart Career Guidance Platform leverages the MERN stack with NLP to provide personalized
career support. It parses user resumes, extracts structured information such as skills education, and
experience, and matches them with relevant job roles using a recommendation engine. The system
highlights missing skills, suggests learning resources, and provides resume optimization feedback for
ATS compatibility by combining data driven insights with user-friendly design, the platform bridges
the gap between individual capabilities and job-market demands.
"""

# 3. Research Analysis Call
try:
    response = client.responses.create(
        model="gpt-4.1",
        input=[
            {
                "role": "system",
                "content": [
                    {"type": "input_text", "text": "You are a Research AI Analyst. Analyze the paper for: Core Thesis, Methodological Innovation, Quantitative Results, and Critical Limitations."}
                ]
            },
            {
                "role": "user",
                "content": [
                    {"type": "input_text", "text": f"Analyze this paper: {research_paper_text}"}
                ]
            }
        ]
    )

    print("--- RESEARCH ANALYSIS REPORT ---")
    print(response.output[0].content)

except Exception as e:
    print(f"Analysis Error: {e}")
```

```python
import json

try:
    # 1. Get the first item in the content list
    # 2. Access the 'text' attribute of that item
    output_text_item = response.output[0].content[0]

    # In some SDK versions, it's a dict; in others, it's an object.
    # This handles both:
    if isinstance(output_text_item, dict):
        raw_json_string = output_text_item.get("text", "")
    else:
        raw_json_string = output_text_item.text

    # Now parse the actual string
    extracted_data = json.loads(raw_json_string)

    print("--- SUCCESS: DATA PARSED ---")
    print(json.dumps(extracted_data, indent=2))

except Exception as e:
    print(f"Parsing Error: {e}")
    # Troubleshooting: print the type to see what exactly was returned
    print(f"Actual content type: {type(response.output[0].content[0])}")
```

```python
from azure.storage.blob import BlobServiceClient

# ✅ Correct connection string
AZURE_STORAGE_CONNECTION_STRING = ""

blob_service_client = BlobServiceClient.from_connection_string(
    AZURE_STORAGE_CONNECTION_STRING
)

container_name = "resumecontainer"

blob_client = blob_service_client.get_blob_client(
    container=container_name,
    blob="Research_analyzer_test.txt"
)

blob_client.upload_blob("Testing Analyzer", overwrite=True)

print("✅ Upload successful")
```
